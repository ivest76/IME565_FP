```python
aqi = pd.read_csv("annual_aqi_by_county_2022.csv")
```

```python
aqi['CO_perc'] = aqi['Days CO'] / aqi['Days with AQI']
aqi['NO2_perc'] = aqi['Days NO2'] / aqi['Days with AQI']
aqi['O3_perc'] = aqi['Days Ozone'] / aqi['Days with AQI']
aqi['PM2.5_perc'] = aqi['Days PM2.5'] / aqi['Days with AQI']
aqi['PM10_perc'] = aqi['Days PM10'] / aqi['Days with AQI']
```

```python
aqi = aqi[['State', 'County','Median AQI', 'CO_perc', 'NO2_perc',
           'O3_perc', 'PM2.5_perc', 'PM10_perc']]
```
